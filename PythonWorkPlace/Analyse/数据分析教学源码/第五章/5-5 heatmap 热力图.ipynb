```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('flights.csv')
```

```python
data.head()
```

```python
data = data.pivot_table(index='month',columns='year',values='passengers')
```

```python
data
```

```python
# 简单看下热力图

# 数值由小到大，颜色从暗到浅
# 一是随着时间的推移，飞机的乘客数量是在逐步增多的
# 二是航班的乘坐旺季在七月和八月份
```

```python
# 设置图片大小

sns.set_context({"figure.figsize":(7,7)})
```

```python
sns.heatmap(data)
```

```python
# cmap
# 设置颜色带的色系
# GnBu
# 常见颜色选择：https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
sns.heatmap(data,cmap='GnBu')
```

```python
# center
# 设置颜色带的分界线
# 颜色带上色彩两级的分界线变成了400
```

```python
sns.heatmap(data,cmap='GnBu',center=400)
```

```python
# annot 
# 是否显示数值注释
# 默认采用科学计数法
# annot=True
```

```python
sns.heatmap(data,cmap='GnBu',center=400,annot=True)
```

```python
# fmt
# 设置数值的格式化形式
# fmt='d'
```

```python
sns.heatmap(data,cmap='GnBu',center=400,annot=True,fmt='d')
```

```python
# linewidths
# 控制每个小方格之间的间距
# linewidths=0.2
```

```python
sns.heatmap(data,cmap='GnBu',center=400,annot=True,fmt='d',linewidths=0.2)
```

```python
# linecolor
# 设置分割线的颜色
# linecolor='grey'
```

```python
sns.heatmap(data,cmap='GnBu',center=400,annot=True,fmt='d',linewidths=0.2)
```
