```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
# stripplot(分布散点图)
```

```python
sns.stripplot(x='time',y='total_bill',data=df)
```

```python
# 利用hue参数进行内部的二次分类
```

```python
sns.stripplot(x='time',y='total_bill',data=df,hue='day')
```

```python
# 把组内的不同类别分开来显示
# dodge
```

```python
sns.stripplot(x='time',y='total_bill',data=df,hue='day',dodge=True)
```

```python
# 对x参数所选字段内的类别进行排序以及筛选
# order
```

```python
sns.stripplot(x='time',y='total_bill',data=df,hue='day',dodge=True
             ,order=['Lunch','Dinner'])
```

```python
# 调整散点图标记点大小
```

```python
sns.stripplot(x='time',y='total_bill',data=df,hue='day',dodge=True
             ,order=['Lunch','Dinner']
             ,size=9)
```

```python
# figsize 调整画布大小
```

```python
figure,axes = plt.subplots(1,1,figsize=(10,10))

sns.stripplot(x='time',y='total_bill',data=df,hue='day',dodge=True
             ,order=['Lunch','Dinner']
             ,size=9
             ,ax=axes)
```
