```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# lmplot是用来绘制回归图的，通过lmplot我们可以直观地总览数值数据的内在关系
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
# 初步看下数据
# 指定 x 与 y，以及 数据源
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,fit_reg=False)
```

```python
# 如果只想显示散点图，设置 fit_reg=False
```

```python
# hue 进一步对数据进行二次分类
# hue = sex
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
# col 在列维度上进行分类
# col='sex'
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',col='sex')
```

```python
# row 在行维度上进行分类
# row='smoker'
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',col='sex',row='smoker')
```

```python
# aspect
# 控制图的长宽比
# aspect=2 
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',col='sex',row='smoker',aspect=2)
```

```python
# 控制进行回归的幂次
# order
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,order=4)
```
