```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# displot 集合了matplotlib的 直方图 与核函数估计曲线kdeplot的功能
# 增加了rugplot分布观测条
```

```python
x = np.random.randn(100)
```

```python
# 简单看下
```

```python
sns.distplot(x)
```

```python
# 通过hist和kde参数调节是否显示直方图及核密度估计
```

```python
# 创建一个一行三列的画布
```

```python
figure,axes = plt.subplots(1,3,figsize=(8,5))

sns.distplot(x,ax=axes[0])
sns.distplot(x,ax=axes[1],hist=False)
sns.distplot(x,ax=axes[2],kde=False)
```

```python
# bins 控制直方图划分
# bins=20 划分成20个区间
```

```python
sns.distplot(x,bins=20)
```

```python
# rug = True 生成数值的小细条
```

```python
sns.distplot(x,bins=20,rug=True)
```

```python
# 扩展 ： 拟合标准正态分布 曲线
```

```python
from scipy.stats import norm
```

```python
sns.distplot(x,bins=20,rug=True,fit=norm)
```
