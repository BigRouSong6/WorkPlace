```python
import seaborn as sns
import pandas as pd
```

```python
# 箱形图（Box-plot）又称为盒须图、盒式图或箱线图，是一种用作显示一组数据分散情况资料的统计图。
# 它能显示出一组数据的最大值、最小值、中位数及上下四分位数。因形状如箱子而得名。
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
# 简单绘制下箱线图
# 查看下 y = 'total_bill' 箱线图分布
```

```python
sns.boxplot(y='total_bill',data=data)
```

```python
# 查看下不同 sex 下，total_bill 箱线图分布
# x='sex',y='total_bill'
```

```python
sns.boxplot(x='sex',y='total_bill',data=data)
```

```python
# 查看下不同 sex time下，total_bill 箱线图分布
# 进一步对类别划分
# hue='time'
```

```python
sns.boxplot(x='sex',y='total_bill',data=data,hue='time')
```

```python
# fliersize，用于指示离群值观察的标记大小
# fliersize = 10
```

```python
sns.boxplot(x='sex',y='total_bill',data=data,hue='time',fliersize=15)
```

```python
# stripplot(分布散点图)
# x='time',y='total_bill'
```

```python
sns.stripplot(x='time',y='total_bill',data=data)
```

```python
# 利用hue参数进行内部的二次分类
# hue = 'day'
```

```python
sns.stripplot(x='time',y='total_bill',data=data,hue='day')
```

```python
# 把组内的不同类别分开来显示
# dodge = True
```

```python
sns.stripplot(x='time',y='total_bill',data=data,hue='day',dodge=True)
```
