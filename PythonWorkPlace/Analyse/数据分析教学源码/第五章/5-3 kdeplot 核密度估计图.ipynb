```python
import seaborn as sns
import numpy as np
```

```python
# 核密度估计(kernel density estimation)是在概率论中用来估计未知的密度函数，属于非参数检验方法之一。
# 通过核密度估计图可以比较直观的看出数据样本本身的分布特征
```

```python
# 随机生成100个符合标准正太分布的样例
```

```python
x = np.random.randn(100)
```

```python
sns.kdeplot(x)
```

```python
# cumulative ：是否绘制累积分布
```

```python
sns.kdeplot(x,cumulative=True)
```

```python
# shade：若为True，则在kde曲线下面的区域中进行阴影处理
```

```python
sns.kdeplot(x,shade=True)
```

```python
# color控制曲线及阴影的颜色
```

```python
sns.kdeplot(x,shade=True,color='black')
```

```python
# vertical = False 代表在横轴绘制
# vertical = True 代表在纵轴绘制
```

```python
sns.kdeplot(x,shade=True,color='black',vertical=True)
```

```python
# 二元kde图像
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
```

```python
sns.kdeplot(x,y,shade=True,color='black')
```

```python
# cbar：参数若为True，则会添加一个颜色棒(颜色帮在二元kde图像中才生效)
```

```python
sns.kdeplot(x,y,shade=True,color='blue',cbar=True)
```
