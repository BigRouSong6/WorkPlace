```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 折线
x_1 = [2011,2012,2013,2014,2015,2016,2017]
y_1 = [58000,60200,63000,71000,84000,90500,107000]
y_2 = [52000,54200,51500,58300,56800,59500,62700]
```

```python
#  条形图
bar_x = ['a','b','c','d']
bar_y = [20, 10, 30, 25]
```

```python
# 直方图
s = np.random.randn(500)
```

```python
figure,axes =  plt.subplots(nrows=1,ncols=2,figsize=(10,5))

axes[0].plot(x_1,y_1)
axes[1].plot(x_1,y_2)

plt.show()
```

```python
figure,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x_1,y_1)
axes[0,1].plot(x_1,y_2)
axes[1,0].bar(bar_x,bar_y)
axes[1,1].hist(s)
```
