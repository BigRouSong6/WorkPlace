```python
# 创建Series

# 通过kind 可更改图片类型 line / hist / bar / kde
# grid 是否设置网格
# label
```

```python
s1 = pd.Series(np.random.randn(500)).cumsum()
```

```python
s1
```

```python
# line
# 横轴为 索引
# 纵轴为 数值
```

```python
plt.plot(s1.index,s1.values)
```

```python
s1.plot(kind='line')
```

```python
# hist
# rwidth
```

```python
s1.plot(kind='hist',rwidth=0.7)
```

```python
# bar
# head 10
```

```python
s1[:20].plot(kind='bar')
```

```python
# barh
```

```python
s1[:20].plot(kind='barh')
```

```python
# box
```

```python
s1.plot(kind='box')
```

```python
# kde 
# 密度曲线
```

```python
s1.plot(kind='kde')
```
