```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [2011,2012,2013,2014,2015,2016,2017]
y_data = [58000,60200,63000,71000,84000,90500,107000]
y_data2 = [52000,54200,51500,58300,56800,59500,62700]
```

```python
# 简单绘制
```

```python
plt.plot(x_data,y_data)
plt.plot(x_data,y_data2)
plt.show()
```

```python
# 指定折线颜色
# color
```

```python
plt.plot(x_data,y_data,color='red')
plt.plot(x_data,y_data2,color='black')
plt.show()
```

```python
# linewidth
# 指定折线宽度
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3)
plt.plot(x_data,y_data2,color='black',linewidth=3)
plt.show()
```

```python
# linestyle 指定折线的样式
# -    代表实线
# --   代表虚线
# -.   代表短线、点相间的虚线
# :    代表点线
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':')
plt.show()
```

```python
# label 
# 指定标签说明
# 需 plt.legend()
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.legend()
plt.show()
```

```python
# xlabel
# 添加x轴描述
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.xlabel('year')
plt.legend()
plt.show()
```

```python
# ylabel
# 添加y轴描述
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.xlabel('year')
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
# title
# 添加标题
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('test')
plt.legend()
plt.show()
```

```python
# axis
# 设置坐标轴范围
# plt.axis([x_min, x_max, y_min, y_max])
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('test')
plt.axis([2013,2016,20000,150000])
plt.legend()
plt.show()
```

```python
# grid
# 添加网格
# axis='x' 在x轴添加网格
# axis='y' 在y轴添加网格
```

```python
plt.plot(x_data,y_data,color='red',linewidth=3,linestyle='--',label='data')
plt.plot(x_data,y_data2,color='black',linewidth=3,linestyle=':',label='data2')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('test')
plt.axis([2013,2016,20000,150000])
plt.grid()
plt.legend()
plt.show()
```
