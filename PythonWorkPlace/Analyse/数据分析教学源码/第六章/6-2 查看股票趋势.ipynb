```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nio   = pd.read_csv('nio.csv')
tesla = pd.read_csv('tesla.csv')
```

```python
nio.head()
```

```python
tesla.head()
```

```python
# 先设置好索引
```

```python
nio.set_index('Date',inplace=True)
```

```python
tesla.set_index('Date',inplace=True)
```

```python
# subplots
```

```python
figure,axes = plt.subplots(1,2,figsize=(10,8))

nio['Adj Close'].plot(ax = axes[0],label='nio')
tesla['Adj Close'].plot(ax = axes[1],label='tesla')

axes[0].legend()
axes[1].legend()

plt.show()
```

```python
# 查看下 每日股票的振幅
# High - Low
# figsize=(20,5)
```

```python
tesla.head()
```

```python
nio['change'] = nio['High'] - nio['Low']
```

```python
tesla['change'] = tesla['High'] - tesla['Low']
```

```python
figure,axes = plt.subplots(1,2,figsize=(10,8))

nio['change'].plot(ax = axes[0],label='nio')
tesla['change'].plot(ax = axes[1],label='tesla')

axes[0].legend()
axes[1].legend()

plt.show()
```

```python
# 收盘价 每日环比
# tesla['Adj Close'].pct_change()
```

```python
tesla['pct_change'] = tesla['Adj Close'].pct_change()
nio['pct_change']   = nio['Adj Close'].pct_change()
```

```python
tesla.head()
```

```python
nio.head()
```

```python
figure,axes = plt.subplots(2,2,figsize=(20,5))
tesla['pct_change'].plot(ax=axes[0,0],label='tesla_line',kind='line')
tesla['pct_change'].plot(ax=axes[0,1],label='tesla_hist',kind='hist')

nio['pct_change'].plot(ax=axes[1,0],label='nio_line',kind='line')
nio['pct_change'].plot(ax=axes[1,1],label='nio_hist',kind='hist')

axes[0,0].legend()
axes[0,1].legend()
axes[1,0].legend()
axes[1,1].legend()
plt.show()
```
