# 1.loc

```python
# loc[row,column]
# row 选择指定行数索引
    # df.loc[:,:]  返回所有数据
    # df.loc[:3,:] 返回前4行 (0,1,2,3)

# column 选择指定的列名
    # df.loc[:,'StockCode']   返回  Series
    # df.loc[:,['StockCode']] 返回  DataFrame
```

```python
df.loc[:3,['StockCode','Description','Quantity']]
```

```python
df.loc[:3,'StockCode']
```

```python
# 筛选特定条件的DataFrame
# 选择 StockCode 为 71053
```

```python
df.head()
```
