# 1.map

```python
data = pd.read_csv('map.csv')
```

```python
data.head()
```

```python
# 如果想将 gender 中 
# 男  --> 1
# 女  --> 0
```

```python
# Series
# map
# arg 
   # dict 根据字典的键值对，替换Series里面的值。
   # 若字典中，没有找到Series中对应的值，则赋值为NaN
# na_action 
  # 如果传入 ignore ,则跳过对空值的操作
```

```python
data.gender.map({'男':1,'女':0})
```

```python
# arg
```

```python
# na_action 
# 如果传入 ignore ,则跳过对空值的操作
```

```python
s = pd.Series(['cat', 'dog', np.nan, 'rabbit'])
```

```python
s
```

```python
s.map('i am a {}'.format,na_action='ignore')
```

# 2.replace

```python

# to_replace，若在Series中存在的值即替换，Series中不存在的值维持原状
# inplace
```

```python
df = pd.DataFrame(
        {
            '名称':['产品1','产品2','产品3','产品4','产品5'],
            '金额':[0,0.48,'F',0.74,0],
            '合计':[0.37,0.28,'C',0.57,0.06],
        }
        )
```

```python
df
```

```python
# 指定列替换
```

```python
df.金额.replace({'F':0},inplace=True)
```

```python
df
```

```python
df.replace('C',2)
```
