```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'lemon'],
                    'value': [1, 2, 3, 5]})

df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'jack'],
                    'value': [5, 6, 7, 8]})
```

```python
df1
```

```python
df2
```

```python
# inner join 内连接
# 根据 on 条件，列出左右两表共有的数据
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='inner',suffixes=('_lkey','_rkey'))
```

```python
# left join  左连接
# 产生表1的完全集，而表2中匹配的则有值，没有匹配的则以null值取代
```

```python
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='left',suffixes=('_lkey','_rkey'))
```

```python
# right join 右连接
# 产生表2的完全集，而表1中匹配的则有值，没有匹配的则以null值取代
```

```python
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='right',suffixes=('_lkey','_rkey'))
```

```python
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='outer',suffixes=('_lkey','_rkey'))
```
