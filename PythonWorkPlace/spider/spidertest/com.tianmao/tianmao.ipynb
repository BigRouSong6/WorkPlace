```python
#发送请求,数据提取
need_data = pd.DataFrame(columns=["商品ID","访客数量","商品名称","店铺","支付卖家数","降低排名数","排名"])
for page in range(1,9):
    print("第"+str(page)+"页")
    datasss = pd.DataFrame(columns=["商品ID","访客数量","商品名称","店铺","支付卖家数","降低排名数","排名"])
    response = get_response(page)
    lists = response["data"]["data"]
    for data in lists:
        info_dicts = {}
        info_dicts["商品ID"] =  data["itemId"]["value"]
        try:
            info_dicts["访客数量"] = data["uv"]["value"]
        except Exception as e:
            print(e)
            info_dicts["访客数量"] = ""
        finally:
            info_dicts["商品名称"] = data["item"]["title"]
            info_dicts["店铺"] = data["shop"]["title"]
            info_dicts["支付卖家数"] = data["payByrCnt"]["value"]
            try:
                info_dicts["降低排名数"] = data["cateRankId"]["cycleCqc"]
            except Exception as e:
                print("第"+str(page)+"页___降低排名数")
                info_dicts["降低排名数"] = "新上榜"
            finally:
                info_dicts["排名"] = data["cateRankId"]["value"]
                info_df = pd.DataFrame(info_dicts,index=[0])
                datasss = pd.concat([datasss,info_df])
    sleep(random.randint(1,3))
    need_data = pd.concat([need_data,datasss])
```

```python
# 数据清洗,优化
# #提取排名变化
def get_rank_change(X):
    try:
        X = -int(X)
    except Exception as e:
        print(e)
        X = "新上榜"
    finally:
        return X

need_data["排名变化"] =need_data.降低排名数.apply(lambda X:get_rank_change(X))
# #提取省份
area=pd.read_excel(r"./省份.xlsx")
def get_province(x):
    try:
        x["省份"]=[sa_name for sa_name in area["sa_name"] if re.search(sa_name,x["商品名称"])] [0]
    except Exception as e:
            print(e)
    finally:
        return x
need_data =need_data.apply(get_province,axis=1)

# #提取面值
face_pd = pd.DataFrame({'面值': ["1元",'10元', '20元', '30元', '50元', '100元', '200元', '300元', '500元'] })
def get_face(x):
    try:
        x["面值"]=[face for face in face_pd["面值"] if re.search(face,x["商品名称"])] [0]
    except Exception as e:
            print(e)
    finally:
        return x
need_data =need_data.apply(get_face,axis=1)

# #提取运营商
isp_pd = pd.DataFrame({'运营商': ['移动', '联通', '电信'] })
def get_face(x):
    try:
        x["运营商"]=[isp for isp in isp_pd["运营商"] if re.search(isp,x["商品名称"])] [0]
    except Exception as e:
            print(e)
    finally:
        return x
need_data =need_data.apply(get_face,axis=1)

# 去除列
need_data = need_data.drop("降低排名数",axis=1)
```
